```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(["How to learn programming Fast?","Github is the best software!", "Why Programming is Hard?"])
y = np.array([5200,3300,7000])
```

```python
x_codes = np.arange(len(x))
```

```python
plt.xlabel("Name of Videos")
plt.ylabel("Frequency")
plt.title("Histogram for different topics")
plt.xticks(x_codes,x, rotation=45)

plt.hist(x_codes, bins=np.arange(len(x)+1),align='left',rwidth=0.5)
plt.savefig("histogram.png",bbox_inches='tight',pad_inches=2,transparent=True)
```
