```python
# Source: https://www.kaggle.com/datasets/nelgiriyewithana/countries-of-the-world-2023
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("world-data-2023.csv")
df.head()
```

```python
describe_df = df.describe()
describe_df.to_csv("describe_df.csv")
pd.read_csv("describe_df.csv")
```

```python
df[df["Birth Rate"] == df["Birth Rate"].max()]
```

```python
df[df["Birth Rate"] == df["Birth Rate"].min()]
```

```python
statistics_birth_rate = df["Birth Rate"].head()
statistics_agricultural_land = df["Agricultural Land( %)"].head()
 
plt.ylabel("Agricultural Land( %)")
plt.xlabel("Birth Rate")
plt.plot(statistics_birth_rate,statistics_agricultural_land,color='#008840',marker='+',linestyle='--',markersize=20,label="Birth Rate and Agricultural Land( %)")

plt.legend(loc='best',shadow=True,fontsize='small')

plt.savefig("Agricultural_Land_and_Birt_Rate.png",bbox_inches="tight",transparent=True)
```

```python
df["Birth Rate"].head()
```

```python
statistics_latitude = df["Latitude"].head()
statistics_longitude = df["Longitude"].head()

plt.ylabel("Value")
plt.xlabel("People")
plt.plot(statistics_latitude,'rD--',label="Latitude")
plt.plot(statistics_longitude,'bD--',label="Longitude")

plt.legend(loc='upper right',shadow=True,fontsize='large')
plt.savefig("latitude_logitude.png", bbox_inches="tight",pad_inches=2,transparent=True)
```
