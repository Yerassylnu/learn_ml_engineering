```python
# Source: https://www.kaggle.com/datasets/alessandrolobello/the-ultimate-film-statistics-dataset-for-ml
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("movie_statistic_dataset.csv")
df.head()
```

```python
described = df.describe()
described.to_csv("describe.csv")
pd.read_csv("describe.csv")
```

```python
statistics_runtime_minutes = df["runtime_minutes"].head()
xpos = np.arange(len(df["movie_title"].head()))
plt.xticks(xpos,df["movie_title"].head(),rotation=45)

plt.plot(statistics_runtime_minutes,'b+--',markersize=20)
plt.savefig("plot.png",bbox_inches="tight",pad_inches=1.3,transparent=True)
```
