```python
# Source: https://www.kaggle.com/datasets/mrcalvinwhite/technology-perception-survey
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset - Factors Influencing Technology Adoption in Consumer Households.csv")
df.head()
```

```python
df_names = np.array(["Frequency_LaptopDesktop","Frequency_Smartphone"])
df_things = np.array([df["Frequency_LaptopDesktop"].head(1),
                      df["Frequency_Smartphone"].head(1)])
df_things = df_things.ravel()

plt.axis("equal")
plt.pie(df_things,labels=df_names,autopct="%0.2f%%",shadow=True,explode=[0.2,0])
plt.savefig("pie.png", bbox_inches="tight",pad_inches=1.5,transparent=True)
```
