```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array(["Как изучить программирование быстро?","Гитхаб это лучший софт!", "Почему Программирование СЛОЖНОЕ?"])
y = np.array([5200,3300,7000])
```

```python
plt.xticks(rotation=45)
plt.xlabel("Название видео")
plt.ylabel("Просмотры")
plt.title("Название и просмотры видео")
plt.bar(x,y)
plt.savefig("bar.pdf",bbox_inches="tight",pad_inches=1.2,transparent=True)
```

```python
df = pd.DataFrame({"Название видео":x,
                   "Просмотры": y})
df
```

```python
max1 = df[df["Просмотры"] == df["Просмотры"].max()]
```

```python
min1 = df[df["Просмотры"] == df["Просмотры"].min()]
```

```python
df2 = pd.concat([max1,min1])
df2
```

```python
df2.to_csv("max_min_values.csv",index=False)
pd.read_csv("max_min_values.csv")
```
