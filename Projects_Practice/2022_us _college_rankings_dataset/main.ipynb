```python
# Source: https://www.kaggle.com/datasets/dylankarmin/2022-college-rankings-compared-to-tuition-costs
import pandas as pd
df = pd.read_csv("2022 US College Rankings.csv")
df.head()
```

```python
describe_df = df.describe()
describe_df.to_csv("describe.csv")
pd.read_csv("describe.csv")
```

```python
max_df = df[df["Adjusted Rank"] == df["Adjusted Rank"].max()]
max_df.to_csv("max_df.csv")
pd.read_csv("max_df.csv")
```

```python
min_df = df[df["Adjusted Rank"] == df["Adjusted Rank"].min()]
min_df.to_csv("min_df.csv")
pd.read_csv("min_df.csv")
```

```python
price_quality_ratio = df["PriceQualityRatio"] = df["Adjusted Rank"] / df["Tuition"]
price_quality_ratio.to_csv("price_quality_ratio.csv")
pd.read_csv("price_quality_ratio.csv")
```
