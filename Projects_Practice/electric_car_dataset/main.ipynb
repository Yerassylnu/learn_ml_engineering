```python
import pandas as pd
df_normal = pd.read_csv("ElectricCarData_Norm.csv")
df_normal.head()
```

```python
df_clean = pd.read_csv("ElectricCarData_Clean.csv")
df_clean.head()
```

```python
new_df = df_clean
new_df = new_df.rename(columns={
    "AccelSec": "Accel",
    "TopSpeed_KmH": "TopSpeed",
    "Range_Km": "Range",
    "Efficiency_WhKm": "Efficiency",
    "FastCharge_KmH": "FastCharge"
})

new_df["TopSpeed"] = new_df["TopSpeed"].astype(str) + " km/h"
new_df["Range"] = new_df["Range"].astype(str) + " km"
new_df["Efficiency"] = new_df["Efficiency"].astype(str) + " Wh/km"
new_df["FastCharge"] = new_df["FastCharge"].astype(str) + " km/h"
new_df.head()
```

```python
new_df.to_csv("new_df.csv")
```
