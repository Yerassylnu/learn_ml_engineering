```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.fillna({
    "Cost $": "not available"
})
new_df = new_df.interpolate()
new_df = new_df.rename(columns={
    "Cost $": "Cost",
    "Speed km/h": "Speed",
    "Stamina /10": "Stamina"
})
new_df["Cost"] = new_df["Cost"].astype(str) + " $"
new_df["Speed"] = new_df["Speed"].astype(str) + " km/h"
new_df["Stamina"] = new_df["Stamina"].astype(str) + " /10"
new_df
```

```python
new_df.to_csv("cars_analyze.csv")
pd.read_csv("cars_analyze.csv")
```
