```python
import pandas as pd
import numpy as np

df = pd.read_csv("my_tiktok.csv",sep=';')
df
```

```python
df.describe()
```

```python
views_max = df[df["Просмотры"]==df["Просмотры"].max()]
likes_max = df[df["Лайки"]==df["Лайки"].max()]
comments_max = df[df["Комментарии"]==df["Комментарии"].max()]
saves_max = df[df["Сохранения"]==df["Сохранения"].max()]

concat_df = pd.concat([views_max, likes_max, comments_max, saves_max], keys=["Просмотры","Лайки","Комментарии","Сохранения"])
concat_df

concat_df.to_csv("max_value_tiktok.csv",sep=';')
```

```python
new_df = pd.read_csv("max_value_tiktok.csv",sep=';')
new_df
```

```python
filtered_df = df[df["Просмотры"] >= 1000]
filtered_df
```

```python
concat_df2 = pd.concat([concat_df.reset_index(drop=True),filtered_df],keys=["Максимальные значения", "Просмотры больше 1000"])
concat_df2
```

```python
concat_df2.to_csv("my_tiktok_analyzed.csv",sep=';')

most_popular = np.array([])

for item in filtered_df["Рубрика"]:
    most_popular = np.append(most_popular,item)

df_most_popular = pd.DataFrame({
    "Популярные темы": most_popular
})

df_unique = df_most_popular.drop_duplicates()
df_unique.to_csv("my_tiktok_results.csv",sep=';',index=False)
    
read_analyzed = pd.read_csv("my_tiktok_analyzed.csv",sep=';')
read_results = pd.read_csv("my_tiktok_results.csv",sep=';')
```
